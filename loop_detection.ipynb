```python
import os
```

```python
def get_push_values(hexnum):
    # break hexadecimal number into a list of 256-bit values
    # hexnum: hexadecimal number in string form
    # rtype: a list of decimal numbers (one corresponds to 2 "bits" in the hex number)
    hex_num = hexnum[2:]
    vals = [int(hex_num[i:i+2],16) for i in range(0, len(hex_num), 2)] 
    return vals
```

```python
def loop_detection(opcode_file):
    
    with open(opcode_file) as f:
        data = f.read()
    data = data.replace('Unknown Opcode', 'UnknownOpcode')
    instrs = data.split()

    
    res = []
    for i in range(len(instrs)):
        if instrs[i] == 'JUMP':
            if instrs[i-2].startswith('PUSH'):
                values = get_push_values(instrs[i-1]) # values to be pushed (as in a list)
                if int(instrs[i-2][4:]) == len(values):
                    dest_pos = values[-1] # the top value in the stack
                else:
                    print('invalid push (does not match #values) at ', i)

                    
                if dest_pos < i and instrs[dest_pos] == 'JUMPDEST':
                    res.append((i, dest_pos))

        elif instrs[i] == 'JUMPI':
            if instrs[i-2].startswith('PUSH'):
                values = get_push_values(instrs[i-1]) # values to be pushed (as in a list)

                if int(instrs[i-2][4:]) == len(values):
                    dest_pos = values[-1] # the top value in the stack
                    
                else:
                    print('invalid push (does not match #values) at ', i)

                    
                if dest_pos < i and instrs[dest_pos] == 'JUMPDEST' and values[-2] != 0:
                    res.append((i, dest_pos))


    if len(res)>0:
        return res
```

```python
res = []
for file in os.listdir("./zchao_dataset/opcode/"):
    filename = os.path.join("./zchao_dataset/opcode/", file)
    res_current = loop_detection(filename)
    if res_current:
        res.append([file,res_current])
```
